```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
file_path = 'Mendelian.train.tsv'
training_dataframe=pd.read_csv(file_path, sep='\t')
```

```python
corr = training_dataframe.corr()
```

```python
plt.rcParams['figure.figsize'] = [20, 20]
plt.axis('scaled')
sns.heatmap(corr, 
        annot=True,
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

```python
dropped = training_dataframe.drop(columns=["CpGobsExp", "CpGperCpG", "dbVARCount", "mamPhyloP46way"])
```

```python
dropped_corr = dropped.corr()
```

```python
plt.rcParams['figure.figsize'] = [20, 20]
plt.axis('scaled')
sns.heatmap(dropped_corr, 
        annot=True,
        xticklabels=dropped_corr.columns,
        yticklabels=dropped_corr.columns)
```
